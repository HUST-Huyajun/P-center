```python
import numpy as np
import pandas as pd
data=pd.read_csv("info.csv")
```

```python
a=data[['k','tabu_step','opt_ans']].groupby(['k','tabu_step']).mean()
a
```

```python
a=a.sort_values(by=['opt_ans'])
gooddata=a[1:20]
gg=gooddata.reset_index().values
#print(gg)
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import math
fig = plt.figure(1)
ax = fig.add_subplot(111)

for yy in gg:
    xs=yy[0]
    ys=yy[1]
    ax.scatter(xs, ys)
    ax.set_xlabel('k')
    ax.set_ylabel('tabu_step')
  

plt.show()
```
